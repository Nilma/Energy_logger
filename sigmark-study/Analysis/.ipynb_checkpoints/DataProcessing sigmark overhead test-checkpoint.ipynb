# The R² Score of the models

R-squared (R²) is defined as a number that tells you how well the independent variable(s) in a statistical model explains the variation in the dependent variable. It ranges from 0 to 1, where 1 indicates a perfect fit of the model to the data.

While R² is often between 0 and 1, it can be negative when the model performs worse than predicting the mean of the data.

```python
from sklearn.metrics import r2_score

r2_dimpact = r2_score(dfs['energy'], dfs['dimpactmodel_energy'])
r2_digst = r2_score(dfs['energy'], dfs['digstmodel_energy'])

print(f"R² for DIMPACT model: {r2_dimpact:.3f}")
print(f"R² for DIGST model: {r2_digst:.3f}")
```

In Conclusion, **both model performs worse than predicting the mean of the sampled data**. In the following, we investigate the models' error. 

# The systematic nature of the models' error

The DIMPACT and DIGST models both assume

$E_{\mathrm{predicted}} = P_{\mathrm{fixed}} \times T_{\mathrm{duration}}$

If the assumed power is incorrect, we can decompose it into the correct mean power plus an error term \(\Delta P\):

$P_{\mathrm{fixed}} = P_{\mathrm{correct}} + \Delta P$

Substituting gives:

$E_{\mathrm{predicted}} = P_{\mathrm{correct}} \times T_{\mathrm{duration}} + \Delta P \times T_{\mathrm{duration}}$

Thus, the energy error ($\Delta E$) grows linearly with time:

$\mathrm{\Delta E} = \Delta P \times T_{\mathrm{duration}}$

From the construction of the model, any bias will appear as small energy errors for short runs, but the same bias is magnified for long runs.

We will demonstrate this using linear regression on the absolute error over the duration of the individual samples. If the best-fitted line has a positive slope it means that the errors grow when time grows.

The positive correlation between time ane absolute error shows that the error grows over duration time.

```python
model = smf.ols("dimpact_abs_error ~ duration + 0 ", data=dfs).fit()
print(model.summary())

# Scatterplot of error vs duration
plt.figure(figsize=(8,6))
sns.scatterplot(x="duration", y="dimpact_abs_error", data=dfs, alpha=0.6, edgecolor=None)

# Regression line (from statsmodels fit)
x_vals = np.linspace(dfs["duration"].min(), dfs["duration"].max(), 200)
y_vals = 0 + model.params['duration']*x_vals  # from the regression model
plt.plot(x_vals, y_vals, color="blue", linewidth=2, label="Fitted line",linestyle='--',)

plt.title("DIMPACT Absolute Error vs Duration")
plt.xlabel("Duration (seconds)")
plt.ylabel("Absolute Error in Energy (Joules)")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig('dimpact_abs_error_vs_duration.png')
plt.show()
```

```python
model = smf.ols("digst_abs_error ~ duration + 0 ", data=dfs).fit()
print(model.summary())

# Scatterplot of error vs duration
plt.figure(figsize=(8,6))
sns.scatterplot(x="duration", y="digst_abs_error", data=dfs, alpha=0.6, edgecolor=None)

# Regression line (from statsmodels fit)
x_vals = np.linspace(dfs["duration"].min(), dfs["duration"].max(), 200)
y_vals = 0 + model.params['duration']*x_vals  # from the regression model
plt.plot(x_vals, y_vals, color="blue", linewidth=2, label="Fitted line",linestyle='--',)

plt.title("Digst Absolute Error vs Duration")
plt.xlabel("Duration (seconds)")
plt.ylabel("Absolute Error in Energy (Joules)")
plt.legend()
plt.grid(True)
plt.grid(alpha=0.3)
plt.savefig('digst_abs_error_vs_duration.png')
plt.show()
```

## Does it improve the results more to consider X (X being computer, category, or flow id) when modelling the energy

### The Best-fit Global Model

$\text{energy}_i = \beta_0 + \beta_1 \cdot \text{duration}_i + \epsilon_i$

### The Best-fit X Model:
$\text{energy}_i = \beta_0 + \beta_1 \cdot \text{duration}_i 
+ \sum_{j} \gamma_j \, I_{\text{X}=j,i} 
+ \sum_{j} \delta_j \, \big(\text{duration}_i \times I_{\text{X}=j,i}\big) 
+ \epsilon_i$


- $\beta_0$: global intercept (baseline energy when duration = 0).
- $\beta_1$: baseline slope (effect of duration).
- $I_{\text{X}=j,i}$ : indicator (1 if observation $i$ is from $X=j$, 0 otherwise).
- $\gamma_j$: adjustment to intercept when $X=j$ 
- $\delta_j$: adjustment to slope (duration effect) for $X=j$  
- $\epsilon_i$: residual error term.

```python

formula_global =   "energy ~ duration+ 0"
formula_computer = "energy ~ duration:C(computerid) - 1"
formula_category =  "energy ~ duration:C(category) -1"
formula_flowID =  "energy ~ duration:C(flowid) -1"
formula_flowID_computer =  "energy ~ duration:C(flowid):C(computerid) -1"
formula_category_computer =  "energy ~ duration:C(category):C(computerid) -1"

formula = formula_global
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t\t\t\t R²:", round(model.rsquared, 3))#, "\tPower: ", round(model.params[0],3))
print(model.summary())

formula = formula_computer
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t\t R²:", round(model.rsquared, 3),"\t Power: ", round(model.params[0],3))
print(model.summary())

formula = formula_category
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t\t R²:", round(model.rsquared, 3)) #, "\tPower: ", round(model.params[0],3))
print(model.summary())

formula = formula_flowID
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t\t\t R²:", round(model.rsquared, 3))#, "\tPower: ", round(model.params[0],3))
print(model.summary())

formula = formula_category_computer
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t R²:", round(model.rsquared, 3)) #, "\tPower: ", round(model.params[0],3))


formula = formula_flowID_computer
model = smf.ols(formula, data=dfs).fit()
print(formula+ "\t R²:", round(model.rsquared, 3))#, "\tPower: ", round(model.params['coef'],3))
```

# What is R²
R-squared (R²) is defined as a number that tells you how well the independent variable(s) in a statistical model explains the variation in the dependent variable. It ranges from 0 to 1, where 1 indicates a perfect fit of the model to the data.

## General interpretation

| R² value       | Rough interpretation         | What it means in practice                                                                                                                                         |
| -------------- | ---------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **0.50 (50%)** | **Moderate / weak**          | Only half of the variation is explained. The model is missing important predictors, or the system is inherently noisy. Predictions may be unreliable.             |
| **0.75 (75%)** | **Good**                     | Three-quarters of the variation is explained. The model captures the main drivers, but there’s still substantial unexplained variance.                            |
| **0.90 (90%)** | **Very good / strong**       | The model explains almost all variation. Predictions are highly reliable. Remaining residuals are small or due to noise.                                          |
| **0.95 (95%)** | **Excellent / near perfect** | Almost everything is explained. Very little residual variation is left — the model fits extremely well, though beware of overfitting if many predictors are used. |




Thus, looking at our results so far:



| Formula                                           | What it means                           | Your R² | Interpretation                                                               |
| ------------------------------------------------- | --------------------------------------- | ------- | ---------------------------------------------------------------------------- |
| `energy ~ duration + 0`                           | One global slope for all runs.          | 0.946   | A single fixed power explains ~95% of variance. Already a strong fit.        |
| `energy ~ duration:C(computerid) - 1`             | Separate slope per computer.            | 0.960   | Hardware differences explain an extra ~1.4% of variance.                     |
| `energy ~ duration:C(category) - 1`               | Separate slope per category.            | 0.973   | Categories explain more (~3% extra). Stronger than computers alone.          |
| `energy ~ duration:C(flowid) - 1`                 | Separate slope per flow.                | 0.978   | Individual flows are even more predictive — ~98% variance explained.         |
| `energy ~ duration:C(category):C(computerid) - 1` | Separate slope per category × computer. | 0.989   | Almost perfect fit — categories behave differently depending on the machine. |
| `energy ~ duration:C(flowid):C(computerid) - 1`   | Separate slope per flow × computer.     | 0.994   | Essentially saturated the variance. Almost all residual error is gone.       |

```python
model = smf.mixedlm("energy ~ duration", data=dfs, groups=dfs["participantid"]).fit()
print(model.summary())

# Extract variance components
between_var = model.cov_re.iloc[0,0]   # variance between users
within_var = model.scale               # residual (within-user) variance
icc = between_var / (between_var + within_var)

print(f"Between-user variance: {between_var:.2f}")
print(f"Within-user variance:  {within_var:.2f}")
print(f"Intra-class correlation (ICC): {icc:.2f}")
```

Between-user variance is tiny (915) compared to within-user variance (74,336).

ICC = 0.01 → only about 1% of the total variance is explained by differences between users (or computers, depending on what you grouped by).

In other words: most of the variation comes from run-to-run noise within the same user/computer, not systematic differences across users.

Adding more users will not help much, because users are already very similar on average.

Adding more repetitions per user is more valuable, because the dominant uncertainty comes from within-user variability.

This fits with what you’ve observed: the model structure (fixed

## Individual participants impact on Total Variance
To quantify the contribution of individual participants to total variance, we fit a random-intercept mixed-effects model with participant ID as the grouping factor. 


```python

    
# Filter out participants with too few observations (<= 2)
valid_ids = dfs["participantid"].value_counts()[dfs["participantid"].value_counts() > 2].index
dfm = dfs[dfs["participantid"].isin(valid_ids)].copy()

# Drop unused category levels and reset everything cleanly
dfm["participantid"] = dfm["participantid"].astype("category")
dfm["participantid"] = dfm["participantid"].cat.remove_unused_categories()
dfm = dfm.reset_index(drop=True)

# Fit a simple random-intercept model 
m = smf.mixedlm("energy ~ duration", data=dfm, groups=dfm["participantid"]).fit()
print(m.summary())

# Compute ICC
user_var = m.cov_re.iloc[0,0]
resid_var = m.scale
icc = user_var / (user_var + resid_var)
print(f"Intraclass Correlation Coefficient (ICC): {icc:.3f}")
```

```python
base_estimates = {}

# Loop over computers
for comp, dfc in df_powers.groupby("computerid"):
    values = dfc["payload"].dropna().values

    # --- Method 1: Histogram bin with max count ---
    counts, bin_edges = np.histogram(values, bins=50)  # adjust bins as needed
    max_bin_idx = np.argmax(counts)
    base_guess = (bin_edges[max_bin_idx] + bin_edges[max_bin_idx+1]) / 2

    # --- Method 2 (optional): Median of lowest 5% of values ---
    low_tail_guess = np.median(np.quantile(values, np.linspace(0, 0.05, 20)))

    base_estimates[comp] = {
        "hist_mode": base_guess,
        "low_tail": low_tail_guess
    }

    # Plot for visual check
    sns.histplot(values, bins=50, kde=False)
    plt.axvline(base_guess, color="red", linestyle="--", label=f"hist_mode={base_guess:.2f}")
    plt.axvline(low_tail_guess, color="green", linestyle="--", label=f"low_tail={low_tail_guess:.2f}")
    plt.title(f"Computer {comp} - Estimated Base Power")
    plt.legend()
    plt.show()

# Show results
for comp, vals in base_estimates.items():
    print(f"Computer {comp}: hist_mode={vals['hist_mode']:.2f}, low_tail={vals['low_tail']:.2f}")
```

```python
base_estimates = {}
xlim = (0, 30)   # adjust if your payload range is wider/narrower


# group by computer and flow
for (comp, flow), df_cf in df_powers.groupby(["computerid", "flowid"]):
    values = df_cf["payload"].dropna().values
    
    # histogram method
    counts, bin_edges = np.histogram(values, bins=50)
    max_bin_idx = np.argmax(counts)
    hist_mode = (bin_edges[max_bin_idx] + bin_edges[max_bin_idx+1]) / 2
    
    # low-tail method (median of lowest 5%)
    low_tail = np.median(np.quantile(values, np.linspace(0, 0.05, 20)))
    
    base_estimates[(comp, flow)] = {"hist_mode": hist_mode, "low_tail": low_tail}
    
    # plot histogram for inspection
    sns.histplot(values, bins=50, kde=False)
    plt.axvline(hist_mode, color="red", linestyle="--", label=f"hist_mode={hist_mode:.2f}")
    plt.axvline(low_tail, color="green", linestyle="--", label=f"low_tail={low_tail:.2f}")
    plt.title(f"Computer {comp}, Flow {flow} - Base power estimate")
    plt.xlim(xlim)
    plt.legend()
    plt.show()

# compact table
import pandas as pd
df_base = pd.DataFrame.from_dict(base_estimates, orient="index")
df_base.index = pd.MultiIndex.from_tuples(df_base.index, names=["computerid","flowid"])
print(df_base)
```
